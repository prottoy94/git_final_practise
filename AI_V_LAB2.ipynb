Basic Arithmetic Operation


```python
a= 20
b=5
sum_r=a+b
sub=a-b
prod=a*b
div=a/b

print("Sum:",sum_r)
print(f"Sum: {sum_r}")
print(f"Sub: {sub}")
print(f"Prod: {prod}")
print(f"Div: {div}")
print()
#or
print(f"Sum: {sum_r}", end=", ")
print(f"Sub: {sub}", end=", ")
print(f"Prod: {prod}", end=", ")
print(f"Div: {div}", end=" ")
```

Power Operation


```python
base=3
exp=3
print(pow(base,exp))  # 2^5
c=7
d=2
#modulus
mod=c%d
div1=c//d
print(f"Result of Modulus: {mod}")  # 7%2
print(f"Result of Floor Division: {div1}")  # 7//2[integer division]
print(f"Result of Floor Division: {c/d}")  # 7/2[float division]
```

String

```python
str1="Hello"
str2="World"
str3="Class"
# Concat
message=str1+" "+str2+" "+str3
print(message)
print(len(message))
# slicing
substr= message[2:7]
print("Sub:", substr)
#find substring
index=message.find("Class")
print("Index of Class:", index)  # returns -1 if not found
#replace
new_message= message.replace("World","Data Structure")
print("New Message:", new_message)
```

```python
#split by whitespace
words= new_message.split(" ")
print(words)
print(len(words))
for word in words:
    print(word, end=" ")
words.append("2025")

print()
print(words)
print(len(words))
for word in words:
    print(word, end=" ")
print()
print(words[0])

print()
words.insert(1, "Engineers")
words
```

Set functions

```python
set(words)
print(set(words))
print(sorted(words))
words.reverse()
print(words)
words.append("Hello")
words.pop(0) #removes from the last
words.pop(0) 
print(words.count("Hello"))
print(words[0].count('l'))
```

LOOP

```python
for i in range(5):
    print(i)

print()
for i in range(2,10):
    print(i)
    
print()
for i in range(2,10,2):
    print(i)

print()
for i in range(10,-1,-1):
    print(i)
```
