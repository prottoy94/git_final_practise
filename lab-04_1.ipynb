```python
def bfs(startNode, goalNode, G):
    visited = []
    expand_list = []
    frontier = []
    
    snode = {'city': startNode, 'Path_Cost': 0, 'Path': [startNode]}
    frontier.append(snode)
    visited.append(startNode)
    
    while len(frontier) > 0:
        print(frontier)
        unode = frontier.pop(0)
        u = unode['city']
        expand_list.append(u)
        print(u, 'expanded')
        
        for v in G[u].keys():
            if v in visited:
                continue
            cost = G[u][v] + unode['Path_Cost']
            path = unode['Path'] + [v]
            vnode = {'city': v, 'Path_Cost': cost, 'Path': path}
            
            if v == goalNode:
                print('Goal node found:', vnode)
                print('Path Cost:', cost)
                return expand_list
            
            frontier.append(vnode)
            visited.append(v)
    
    return expand_list

bfs('Arad', 'Bucharest', romanian_map)
def dfs(startNode, goalNode, G):
    visited=[]
    expand_list=[]
    frontier=[]
    
    snode={'city': startNode, 'Path_Cost': 0, 'Path': [startNode]}
    frontier.append(snode)
    visited.append(startNode)
    
    while len(frontier)>0:
        print(frontier)
        unode=frontier.pop()
        u=unode['city']
        expand_list.append(u)
        print(u, 'expanded')
        
        for v in G[u].keys():
            cost=G[u][v]+unode['Path_Cost']
            path=unode['Path']+[v]
            unode={'city': v, 'Path_Cost': cost, 'Path': path}
            if v==goalNode:
                print('Goal node found:', unode)
                return expand_list
            if v not in visited:
                frontier.append(unode)
                visited.append(v)
    return expand_list
    
    
    
    
    
    
    
    
    
    
    
    
```
